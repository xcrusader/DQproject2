```python
from csv import reader

opened_file = open('hacker_news.csv')
read_file = reader(opened_file)
hn = list(read_file)
```

```python
print(hn[:5])
```

```python
header = hn[0]
hn = hn[1:]

for row in hn[:100]:
    print(row,'\n')
# print(hn[:100])
```

```python
ask_posts = []
show_posts = []
other_posts = []

for row in hn:
    title = row[1]
    if title.lower().startswith('ask hn'):
        ask_posts.append(row)
    elif title.lower().startswith('show hn'):
        show_posts.append(row)
    else:
        other_posts.append(row)
        
print("{0} posts relate to asking".format(len(ask_posts)))
print("{0} posts relate to showing".format(len(show_posts)))
print("{0} posts relate to neither showing or asking".format(len(other_posts)))
```

```python
total_ask_comments = 0
total_show_comments = 0
for row in ask_posts:
    total_ask_comments+=int(row[4])

avg_ask_comments = total_ask_comments / len(ask_posts)
print("The average number of ask related comments is {0}".format(avg_ask_comments))

for row in show_posts:
    total_show_comments+=int(row[4])
    
avg_show_comments = total_show_comments / len(show_posts)
print("The average number of show comments is {0}".format(avg_show_comments))

```

As can be seen ask related comments have a clearly greater average that show related comments due to the nature of questioning. People are usually more inclined to answer a question to not only display their knowledge on the topic but to also know whether their answer will encourage further discussion. 

```python
import datetime as dt
result_list = []

for row in ask_posts:
    temp_list = []
    created_at = row[6]
    number_of_comments = row[4]
    
    temp_list.append(created_at)
    temp_list.append(number_of_comments)
    result_list.append(temp_list)
    
print(result_list)
```

```python
counts_by_hour = {}
comments_by_hour = {}

for row in result_list:
    dt_object = dt.datetime.strptime(row[0], '%m/%d/%Y %H:%M')
    hour = dt_object.strftime('%H')
    number_of_comments = int(row[1])
    if hour in counts_by_hour:
        counts_by_hour[hour]+=1
        comments_by_hour[hour]+=number_of_comments
    else:
        counts_by_hour[hour]=1
        comments_by_hour[hour]=number_of_comments

print(counts_by_hour,'\n')
print(comments_by_hour)
```

```python
avg_by_hour = []

for hour in counts_by_hour:
    avg_by_hour.append([hour, comments_by_hour[hour]/counts_by_hour[hour]])
    
print(avg_by_hour)
```

```python
swap_avg_by_hour = []

for row in avg_by_hour:
    temp_list = []
    temp_list.append(row[1])
    temp_list.append(row[0])
    swap_avg_by_hour.append(temp_list)

print(swap_avg_by_hour)
```

```python
sorted_swap = sorted(swap_avg_by_hour, reverse = True)

print("Top 5 Hours for Ask Posts Comments")
for row in sorted_swap[:5]:
    dt_object = dt.datetime.strptime(row[1], '%H')
    hour = dt_object.strftime('%H:%M')
    print("{0}: {1:.2f} average comments per post".format(hour, row[0]))
```
